```python
def write_list_to_file(lst, filepath):   
  with open(filepath, 'w') as fp:
    for item in lst:
        # write each item on a new line
        fp.write("%s\n" % item)

          
```

```python
import json

def load_jsonl(input_path) -> list:
    """
    Read list of objects from a JSON lines file.
    """
    data = []
    with open(input_path, 'r', encoding='utf-8') as f:
      for line in f:
        data.append(json.loads(line))
    
    print('Loaded {} records from {}'.format(len(data), input_path))
    return data
```

```python
dataset = load_jsonl("/content/drive/My Drive/formatted_alt.txt");
```

```python
import pandas as pd

tweet_texts = []

for d in dataset:
  tweet_texts.append(d.get('data').get('text'))
        
```

```python
filter_data = [x for x in tweet_texts if not x.startswith('RT')]
len(filter_data)
```

```python
write_list_to_file(filter_data, r'/content/drive/My Drive/formatted_tweets.txt')
```

```python
import re
def delete_tokens(infile, outfile, regex):
  with open(infile) as fin, open(outfile, "w+") as fout:
      for line in fin:
        iter = re.finditer(regex, line)
        for i in iter:
          word_omit = i.group()
          print(word_omit)
          line = line.replace(word_omit, "")
        fout.write(line)      
```

```python
delete_tokens("/content/drive/My Drive/final_tweets_cleaned.txt", "/content/drive/My Drive/final2_tweets_cleaned.txt", r'#\w+')
```
